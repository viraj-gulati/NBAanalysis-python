```python
on_street_name = df['ON STREET NAME']
#remove missing vals
on_street_name = on_street_name.dropna()
#counts of streetname
street_counts = on_street_name.groupby(on_street_name).count()

#sort data descending
street_counts = street_counts.sort_values(ascending=False)

#print top 3
print(street_counts.head(3))
```

```python
borough = df['BOROUGH']

#include missing values in other
borough = borough.fillna('Other')

#specify which boroughs we want
borough[~borough.isin(['MANHATTAN', 'BROOKLYN', 'QUEENS', 'BRONX'])] = 'Other'

#group data by count
borough_counts = borough.groupby(borough).count()

#plot
fig, ax = plt.subplots()
ax.pie(borough_counts, labels=borough_counts.index, autopct='%1.1f%%')
ax.set_title('Number of Accidents by Borough')
plt.show()
```

```python
injuries = df[['BOROUGH', 'NUMBER OF PERSONS INJURED']]

#replace misssing valyes
injuries['NUMBER OF PERSONS INJURED'] = injuries['NUMBER OF PERSONS INJURED'].fillna(0)

#summary stats for nyc
all_stats = injuries['NUMBER OF PERSONS INJURED'].describe()

#summary stats for two boroughs. 
borough_stats = injuries[injuries['BOROUGH'].isin(['MANHATTAN', 'BROOKLYN'])].groupby('BOROUGH')['NUMBER OF PERSONS INJURED'].describe()

# Print the results
print("Summary Statistics for All of NYC:\n", all_stats)
print("\nSummary Statistics for Selected Boroughs:\n", borough_stats)
```

On average for every accident in New York City, about 3.6/10 people are injured or .36. Accidens in Brooklyn are lower than the average for the whole city. Meanwhile brooklyn is nearly equal to the average for NYC. 

```python
locations = df[['BOROUGH', 'LATITUDE', 'LONGITUDE']]

# Replace missing values with 'UNKNOWN'
locations['BOROUGH'] = locations['BOROUGH'].fillna('UNKNOWN')
locations['LATITUDE'] = locations['LATITUDE'].fillna(0)
locations['LONGITUDE'] = locations['LONGITUDE'].fillna(0)

# Create a scatter plot to show the accidents by borough
colors = {'BRONX': 'red', 'BROOKLYN': 'blue', 'MANHATTAN': 'green', 'QUEENS': 'orange', 'UNKNOWN': 'gray', 'STATEN ISLAND': 'purple'}
fig, ax = plt.subplots()
for borough in ['BRONX', 'BROOKLYN', 'MANHATTAN', 'QUEENS', 'STATEN ISLAND']:
    if borough != 'UNKNOWN':
        borough_locations = locations[locations['BOROUGH'] == borough]
        ax.scatter(borough_locations['LATITUDE'], borough_locations['LONGITUDE'], c=colors[borough], label=borough, alpha=0.5, s=1)
ax.legend()
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_title('Accidents by Geo Location and Borough')

ax.set_ylim(-73 , -75)
ax.set_xlim(40.4 , 41)# set the y-axis limits
plt.show()
```

```python
from calendar import month_abbr
import calendar

locations = locations[locations['CRASH DATE'].notnull()]

# get the month column by converting the date column to datetime format
locations['CRASH DATE'] = pd.to_datetime(locations['CRASH DATE'])
locations['MONTH'] = locations['CRASH DATE'].dt.month

# count the number of accidents for each month
month_counts = locations['MONTH'].value_counts()

# remove months with no accidents
month_counts = {k: v for k, v in month_counts.items() if k in range(1, 9)}

# get the abbreviated month names
month_names = [month_abbr[i] for i in month_counts.keys()]

# create a pandas DataFrame from the dictionary
df_month_counts = pd.DataFrame({'Month': month_names, 'Accidents': month_counts.values()})

# plot the bar chart using pandas plot method
df_month_counts.plot(kind='bar', x='Month', y='Accidents', legend=False)
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.title('Number of Accidents by Month')
plt.show()
```

it appear as the year progresses, the accidents seem to decline. This lull may be due to the lack of people outside due to the covid-19 virus habing been spread.
